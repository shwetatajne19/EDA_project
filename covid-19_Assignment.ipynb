```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Admin\Desktop\EDA_shweta/covid_india_sample (2).csv')
df.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(7,7))
plt.barh(df['date'],df['mobility_change'],color='green')
plt.title("date vs mobility_change of Covid-19")
plt.xlabel("mobility_change")
plt.ylabel("date")
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(9,9))
plt.bar(df['vaccination_rate'],df['new_cases'],color='orange')
plt.title("vaccination_rate vs new_cases of Covid-19")
plt.xlabel("vaccination_rate")
plt.ylabel("new_cases")
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(7,7))
plt.scatter(df['recovery_rate'],df['date'],color='red')
plt.title("date vs recovery_rate of Covid-19")
plt.xlabel("recovery_rate")
plt.ylabel("date")
plt.show()
```

```python
from matplotlib import pyplot as plt

df.plot(x="date", y=['new_cases','new_deaths'], kind="barh")                     
plt.title("new_cases & new_deaths of Covid-19")
plt.xlabel("count")
plt.ylabel("date")
plt.show()
```
